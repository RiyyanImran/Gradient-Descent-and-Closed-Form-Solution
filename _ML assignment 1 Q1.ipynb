# computing hypothesis

```python
theta=[0.01,0.02,0.03,0.04]

def hypothesis(X,theta):
    X["predictions/hypothesis"] = theta[0]*X["bias"]+theta[1]*X["x1"]+theta[2]*X["x2"]+theta[3]*X["x3"]

    X["error"] = X["predictions/hypothesis"] - X["y"]

    return X
    
hypo=hypothesis(X,theta)
X
```

# Computing Cost function

```python
def cost_function(X):
    X["cost"]=(1/(2*m))*np.sum((X["error"]**2))
    return X
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Load data
X = np.loadtxt("DataX.dat")
Y = np.loadtxt("DataY.dat")

# Extract features
living_area = X[:, 0]
bedrooms = X[:, 1]
floors = X[:, 2]

# Create a 3D scatter plot
fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(111, projection='3d')

# Scatter plot with color gradient based on house prices
sc = ax.scatter(living_area, bedrooms, floors, c=Y, cmap='viridis', marker='o', alpha=0.7)

# Customize the axes labels
ax.set_xlabel('Living Area (sq ft)', fontsize=12)
ax.set_ylabel('Number of Bedrooms', fontsize=12)
ax.set_zlabel('Number of Floors', fontsize=12)
ax.set_title('House Prices vs. Features - 3D Scatter Plot', fontsize=16)

# Set the viewing angle
ax.view_init(elev=20, azim=30)

# Add a color bar to show the price range
cbar = plt.colorbar(sc)
cbar.set_label('House Prices ($)', fontsize=12)

# Customize the grid and ticks
ax.grid(True)
ax.tick_params(axis='both', labelsize=10)

# Show the plot
plt.tight_layout()
plt.show()
```
